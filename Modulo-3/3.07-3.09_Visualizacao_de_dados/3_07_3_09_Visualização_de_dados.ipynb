### Plot de linha

Um dos m√©todos mais utilizados do pyplot (aqui chamado de `plt`) √© o m√©todo `plot`, que permite a cria√ß√£o de gr√°ficos de linha a partir de valores fornecidos para o eixo x e y, que representam os **pontos a serem conectados pela linha do gr√°fico**.

Ao invocar `plt.plot(x, y)`, estamos criando uma nova figura com os dados `x` no eixo horizontal e `y` no eixo vertical, apresentados visualmente como uma linha. Gra√ßas ao comando m√°gico utilizado anteriormente, figuras produzidas em uma c√©lula ser√£o exibidas ao final, com ou sem invocar `plt.show()`. Por√©m, ao remover essa chamada, o Jupyter assume o comportamento padr√£o de apresentar o objeto de retorno da √∫ltima instru√ß√£o contida na c√©lula.


````python
# Plotando uma senoide
x = np.arange(0, 3 * np.pi, 0.2)
y = np.sin(x)
plt.plot(x, y)
plt.show()
````

**Propriedades do plot de linha**

Voc√™ pode transform√°-lo em uma linha tracejada, ou at√© incluir s√≠mbolos que demarcam cada ponto `(x, y)` das entradas, al√©m da linha que os conecta.

* Marcadores: https://matplotlib.org/stable/api/markers_api.html
* Estilos de linha: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
* Lista completa de propriedades de linha: https://matplotlib.org/stable/tutorials/pyplot.html#controlling-line-properties

````python
dados = np.array([[5., 7., 9., 8.],  [14., 20., 23., 27.], [25., 22., 15., 12.]])
X = np.arange(dados.shape[1])

marcadores = ['s', 'o', 'v']
linhas = ['--', ':', '-.']
for i in range(0, dados.shape[0]):
	plt.plot(
		   X, dados[i],
		   marker=marcadores[i],
		   linestyle=linhas[i])
````

### M√∫ltiplos plots e propriedades

Ao invocar m√∫ltiplas vezes algum m√©todo que produz figuras (como o `plt.plot()`), todas as visualiza√ß√µes s√£o sobrepostas na mesma figura, facilitando a compara√ß√£o visual entre diversas fun√ß√µes ou dados. Esse √© o comportamento padr√£o do matplotlib, mas voc√™ tem a alternativa de criar expl√≠citamente m√∫ltiplas figuras (veremos a seguir).

√â poss√≠vel personalizar o gr√°fico para melhor representar os dados e facilitar a interpreta√ß√£o visual. A seguir apresentamos as principais propriedades que podemos personalizar em um gr√°fico.

* `plt.xlabel('r√≥tulo do eixo x')`: Define o r√≥tulo do eixo x.
* `plt.ylabel('r√≥tulo do eixo y')`: Define o r√≥tulo do eixo y.
* `plt.title('Seno, cosseno e 0.01x2')`: Define o t√≠tulo do gr√°fico.
* `plt.legend(['Seno', 'Cosseno', '0.01x2'], loc='upper right')`: Adiciona uma legenda ao gr√°fico, com a localiza√ß√£o especificada no canto superior direito.
* `plt.text(2.55, 1., r'Minhas fun√ß√µes')`: Adiciona texto em uma posi√ß√£o espec√≠fica do gr√°fico, no ponto (2.55, 1.0).
* `plt.axis([-1, 11, -1.2, 1.8])`: Define os limites dos eixos x e y (x0, x1, y0, y1).
* `plt.grid(True)`: Ativa a grade no gr√°fico para melhor visualiza√ß√£o e leitura dos dados.

````python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_001x2 = 0.01 * x**2

# M√∫ltiplos plots inclu√≠dos na mesma figura (comportamento padr√£o)
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.plot(x, y_001x2)

# Principais propriedades
plt.xlabel('rotulo do eixo x')
plt.ylabel('rotulo do eixo y')
plt.title('Seno, cosseno e 0.01x2')
plt.legend(['Seno', 'Cosseno', '0.01x2'], loc='upper right')
plt.text(2.55, 1., r'Minhas fun√ß√µes') # x, y, texto
plt.axis([-1, 11, -1.2, 1.8]) #x0, x1, y0, y1
plt.grid(True)
plt.show()
````

### `figure()`, gr√°fico de barras e cores

Voc√™ acabou de ver m√∫ltiplos plots na mesma figura. E se eu quiser cada plot em uma figura? Bom, basta criar uma nova figura com o comando a seguir. A partir deste ponto, novas chamadas de plot ser√£o feitas na nova figura. A propriedade opcional `figsize` permite definir a largura e altura da visualiza√ß√£o (isso deve ser feito com cautela pois pode alterar a percep√ß√£o sobre os dados).
```python
plt.figure(figsize=(w, h))
```

Criaremos de exemplo um **gr√°fico de barras**. O comando `plt.bar(x, y)` produz um gr√°fico de barras **vertical**, onde `x` representa as categorias ou r√≥tulos das barras, e `y` representa os valores correspondentes a cada categoria. J√° o comando `plt.barh(y, x)` cria um gr√°fico de barras **horizontal**, invertendo a disposi√ß√£o dos eixos: `y` representa as categorias ou r√≥tulos das barras e `x` os valores correspondentes.

```python
plt.bar(x, y) # vertical
plt.barh(y, x) # horizontal
```

Nessa c√©lula tamb√©m, personalizamos a cor dos elementos que representam os nossos dados, nesse caso, as barras. Isso √© feito atrav√©s do par√¢metro `color`, que pode ser uma √∫nica representa√ß√£o de cor (a qual ser√° atribu√≠da a todas as barras) ou uma lista de cores, uma para cada elemento visual (uma por barra). As cores podem ser:

* Cores padr√£o: um √∫nico caractere representando as cores base da biblioteca.
```python
{ 'b': blue, 'g': green, 'r': red, 'c': cyan, 'm': magenta, 'y': yellow, 'k': black, 'w': white }
```

* Cores hexadecimal: C√≥digos hexadecimais na forma `#RRGGBB`, onde `RR`, `GG`, e `BB` s√£o valores hexadecimais que representam a intensidade de vermelho, verde e azul, respectivamente. (Lembre-se: hexadecimal √© a base num√©rica com `16` valores, onde cada n√∫mero pode variar de `0` a `F`, equivalente a `0` a `15` em decimal).
```
	#FF7F50
```

* Tupla de cor: Tupla `(r, g, b)` que especifica as intensidades de vermelho, verde e azul, cada uma variando de 0 a 1.
```
(1, 0.5, 0.5)
```

````python
data = [5., 25., 50., 20.]
names= ['Valor 1', 'Valor 2', 'Valor 3', 'Valor 4']
plt.bar(names, data, color='k')

plt.figure(figsize=(6, 3)) #figsize(largura, altura) em polegadas
plt.barh(names, data, color=['r', 'g', 'b', (1, 0.5, 0.5)])

plt.figure(figsize=(6, 3)) #figsize(largura, altura) em polegadas
plt.barh(range(len(data)), data, color='#FF7F50')
````

### Eixos
O objeto da figura, no entanto, pode ser algo muito mais complexo. A figura √© toda a √°rea onde os elementos gr√°ficos s√£o desenhados, como uma folha de papel em branco onde voc√™ vai desenhar seus gr√°ficos. Voc√™ pode ajustar as propriedades da figura inteira, como seu tamanho e resolu√ß√£o.

Al√©m de m√∫ltiplas figuras, podemos trabalhar tamb√©m com m√∫ltiplos **eixos**(*axes*). Apesar de voc√™ conhecer a palavra "eixo" como as dimens√µes de um √∫nico gr√°fico, **aqui ela representa subplots**, ou seja, cada eixo √© uma subfigura onde os dados podem ser plotados. Os eixos possuem seus pr√≥prios r√≥tulos, marca√ß√µes de grade, t√≠tulo, etc.

> O conceito de eixo no escopo de subplots do Matplotlib est√° relacionado ao conceito de "[pequenos m√∫ltiplos](https://en.wikipedia.org/wiki/Small_multiple)" de visualiza√ß√£o de dados, quando separamos os dados em subgrupos para comparar as estat√≠sticas de cada subgrupo.

Podemos criar uma figura com m√∫ltiplos eixos atrav√©s do comando a seguir, que retorna n√£o s√≥ a refer√™ncia √† figura inteira, como tamb√©m ao array de eixos.
```python
fig, axs = plt.subplots(nrows=1, ncols=3, sharex=False, sharey=False, figsize=(15, 4))
```
* `nrows=1, ncols=3`: cria uma grade de 1 linha e 3 colunas, resultando em tr√™s eixos (subplots) lado a lado.
* `sharex=False, sharey=False`: indica que os eixos x e y n√£o s√£o compartilhados entre os subplots. Ou seja, cada eixo ter√° seus pr√≥prios limites em x e em y.
* `figsize=(15, 4)`: define o tamanho da figura inteira (todos os 3 eixos) como 15 polegadas de largura e 4 polegadas de altura.

Note que as defini√ß√µes de propriedade a partir de um eixo (e n√£o mais a partir da figura inteira) s√£o m√©todos de nomes distintos, como

```python
axs[0].set_title('T√≠tulo')
```

Quando trabalhamos com o plot comum, com um √∫nico eixo, usamos o equivalente:
```python
plt.title('T√≠tulo')
```

````python
data = [5., 25., 50., 20.]
names= ['Valor 1', 'Valor 2', 'Valor 3', 'Valor 4']

fig, axs = plt.subplots(nrows=1, ncols=3, sharex=False, sharey=False, figsize=(15, 4))
axs[0].bar(names, data)
axs[1].barh(names, data, color='r')
axs[2].barh(range(len(data)), data, color='#FF7F50')

# muda como definir as propriedades
axs[0].set_title('Figura Vertical com r√≥tulos')
axs[1].set_title('Figura Horizontal com r√≥tulos')
axs[2].set_title('Figura Horizontal com √≠ndices num√©ricos')
plt.show()
````

A fun√ß√£o `plt.subplot` √© uma abordagem mais antiga que permite adicionar subplots manualmente, especificando a posi√ß√£o de cada subplot dentro da grade.  Cada subplot √© adicionado especificando a posi√ß√£o com tr√™s n√∫meros: n√∫mero total de linhas da figura, n√∫mero total de colunas e √≠ndice do subplot que desejamos popular. Esses √≠ndices variam de `1` ao `N` desejado.

````python
fig = plt.figure(figsize=(15, 4))

ax0 = plt.subplot(131) # total de linhas, total de colunas e √≠ndice atual
ax0.bar(names, data)

ax1 = plt.subplot(132) # total de linhas, total de colunas e √≠ndice atual
ax1.barh(names, data)

ax2 = plt.subplot(133) # total de linhas, total de colunas e √≠ndice atual
ax2.barh(range(len(data)), data)
````

### M√∫ltiplas barras

O c√≥digo abaixo cria um gr√°fico de m√∫ltiplas barras, ou seja, temos 3 atributos que desejamos comparar na mesma figura dispondo as barras lado a lado. Para evitar sobreposi√ß√µes, basta manipular 2 atributos:
* Deslocar o eixo x de cada plot
* Alterar a largura da barra (garantindo que a soma de larguras seja `< 1`)

**Propriedade `xticks`**<br>
Neste caso, desejamos um √∫ltimo r√≥tulo para cada grupo de barras, isso pode ser feito personalizando o atributo `xticks`. *Ticks* pode ser livremente traduzido do ingl√™s como "marca√ß√£o", `xticks` portanto s√£o as marca√ß√µes do eixo x.

```python
plt.xticks(np.arange(0, 4)+0.25, colunas)
```
* `np.arange(0, 4) + 0.25`: Calcula as posi√ß√µes dos r√≥tulos. O deslocamento de 0.25 √© adicionado para centralizar os r√≥tulos sob os grupos de barras.
* `colunas`: Os r√≥tulos de texto que aparecem no eixo x.

````python
dados = np.array([[5., 7., 9., 8.],  [14., 20., 23., 27.], [25., 22., 15., 12.]])
X = np.arange(4)
plt.bar(X + 0.00, dados[0], color = 'b', width = 0.267)
plt.bar(X + 0.267, dados[1], color = 'g', width = 0.267)
plt.bar(X + 0.533, dados[2], color = '#FF7F50', width = 0.267)

# Propriedades
colunas = ['2013', '2014', '2015', '2016']
plt.xlabel('r√≥tulo do eixo x')
plt.ylabel('r√≥tulo do eixo y')
plt.title('gr√°fico de multiplas barras')
plt.legend(['Legenda 1', 'Legenda 2', 'Legenda 3'])
plt.xticks(np.arange(0, 4)+0.25, colunas)
plt.show()
````

````python
# Fazendo com la√ßos
X = np.arange(dados.shape[1])
cores = ['b', 'g', '#FF7F50']
compr = 0.8/(dados.shape[0])
for i in range(0, dados.shape[0]):
    plt.bar(X + compr*i, dados[i], color=cores[i], width=compr)
plt.show()
````

### Gr√°fico de Dispers√£o

O gr√°fico de dispers√£o, ou scatter plot, √© uma ferramenta gr√°fica utilizada para visualizar a rela√ß√£o entre duas vari√°veis quantitativas. No Matplotlib, o comando plt.scatter(x, y) √© usado para criar este tipo de gr√°fico, onde x e y representam as vari√°veis nos eixos horizontal e vertical, respectivamente. Cada ponto no gr√°fico representa uma observa√ß√£o do conjunto de dados, permitindo identificar padr√µes, tend√™ncias, correla√ß√µes e outliers entre as duas vari√°veis.

**Tamanho, Cores e colormaps**<br>
No comando a seguir, definimos al√©m dos dados, os seguintes par√¢metros:
* O par√¢metro `c` (*color*) permite definir as cores dos pontos no gr√°fico. Pode ser um √∫nico valor v√°lido de cor, aplicando a mesma cor a todos os pontos, ou uma s√©rie de valores espec√≠ficos para cada ponto. O argumento `c=df['target']` usa os valores da coluna `target` para determinar a cor de cada ponto. Isso facilita a visualiza√ß√£o desse atributo como uma terceira dimens√£o, e como ele varia em rela√ß√£o √†s duas vari√°veis principais.
* O atributo `cmap` (*colormap*) especifica o mapeamento de cores que ser√° utilizado. O argumento `cmap='coolwarm_r'` aplica o colormap chamado `coolwarm_r` (`_r` indica a vers√£o invertida).
    * Colormaps dispon√≠veis: https://matplotlib.org/stable/users/explain/colors/colormaps.html
* O atributo `s` (*size*) define o tamanho dos pontos no gr√°fico de dispers√£o. Pode ser um √∫nico n√∫mero, aplicando o mesmo tamanho a todos os pontos, ou uma s√©rie de valores, onde cada valor especifica o tamanho de um ponto individualmente.

Explore [na documenta√ß√£o](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) outras propriedades do gr√°fico de dispers√£o.

````python
fig = plt.scatter(df['mean radius'], df['mean area'],
            ### cor         # tamanho
            c=df['target'], s=df['worst area']/10,
            cmap='coolwarm_r', edgecolors='k')

plt.legend(handles=fig.legend_elements()[0], labels=['Maligno', 'Benigno'])
````

### `imshow()`

### FacetGrids

Quando definimos um dos atributos `row` ou `col` para produzir subplots, o resultado √© um objeto do tipo `seaborn.axisgrid.FacetGrid`. Esse objeto facilita a cria√ß√£o e a manipula√ß√£o de m√∫ltiplos gr√°ficos (plots) organizados em uma estrutura de grade (grid). Podemos invocar expl√≠citamente esse tipo de dado.
* Tutorial "[Construindo estruturas de multi-plot em grid](https://seaborn.pydata.org/tutorial/axis_grids.html)"

Ao definir o `FacetGrid`, j√° informamos a base de dados utilizada para dar suporte √† defini√ß√£o dos atributos que definem os subconjuntos do grid. No exemplo a seguir, definimos o grid a partir dos dados de pinguim, separando as visualiza√ß√µes em colunas a partir do atributo `species`.

Com o m√©todo `map` do objeto `FacetGrid` criado, podemos aplicar fun√ß√µes de plotagem (como `sns.histplot`) a cada subset de dado.
```
FacetGrid.map(func, *args, **kwargs)
```
* O primeiro par√¢metro `func` √© o m√©todo de plotagem desejado
* Depois, temos um conjunto de par√¢metros posicionais `*args` que devem ser nomes de colunas do DataFrame fornecido ao `FacetGrid`,
* Por fim, os `*kwargs` (*keyword arguments*) s√£o atributos nomeados direcionados para o m√©todo de plotagem (quais par√¢metros aquele plot aceita?)

````python
g = sns.FacetGrid(penguins, col="species")
g.map(sns.histplot,     # func
      "bill_length_mm", # args
      kde=True          # kwargs
)
````

Atributos row, col e hue, al√©m de propriedades da figura

````python
g = sns.FacetGrid(penguins, col="island", hue="species",
                height=4, aspect=.8, sharex=True, sharey=True)
g.map(sns.barplot,               # func
      "body_mass_g", "species",  # args
      width=0.5, orient='h'      # kwargs
)
````

````python
import plotly.express as px
import plotly.graph_objects as go

sa_countries = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia',  'Peru',
                'Ecuador', 'Paraguay', 'Uruguay', 'Venezuela']

df = px.data.gapminder()
df
````

#### Plotly Express
Neste exemplo, com uma √∫nica chamada de fun√ß√£o do Plotly Express, voc√™ pode gerar um gr√°fico de linha para os pa√≠ses Brasil, Uruguai e Peru, colorindo automaticamente as linhas por pa√≠s e adicionando um t√≠tulo ao gr√°fico. > Tire um momento pra fu√ßar o gr√°fico e ver as intera√ß√µes nativamente dispon√≠veis (mouse hover, zoom, pan, clique na legenda, sele√ß√£o de √°rea, etc...)

````python
fig = px.line(df[df.country.isin(sa_countries)],
            x="year", y="lifeExp", color='country',
            title='Expectativa de vida ao longo dos anos'
)
fig.show()
````

#### Plotly Graph Objects
Com Graph Objects, voc√™ cria um gr√°fico de linha para os mesmos pa√≠ses, mas com maior controle sobre o processo. Adiciona tra√ßos separadamente para cada pa√≠s usando um loop, permitindo personaliza√ß√µes espec√≠ficas para cada tra√ßo, como estilos de linha e cores. Ap√≥s adicionar os tra√ßos, voc√™ pode atualizar o layout do gr√°fico para ajustar o t√≠tulo, eixos, e outros elementos de forma mais granular.

Note que, ao criar o gr√°fico trabalhamos apenas com os objetos `pd.Series` do DataFrame, que internamente s√£o arrays Numpy.


````python
fig = go.Figure() # cria a figura
for country in sa_countries: # pra cada linha que eu quero plotar

    df_ = df[df.country == country] # filtra por pa√≠s
    # adiciona uma nova linha
    fig.add_trace(
        # px.line
        go.Scatter(x=df_['year'], y=df_['lifeExp'], name=country)
    )

# defini√ß√£o de propriedades
fig.update_layout(title='Expectativa de vida ao longo dos anos')

fig.show()
````

#### Por que usar Graph Objects ent√£o?

Fa√ßa um experimento: encontre o Brasil no gr√°fico anterior. Ainda que a gente melhore o esquema de cores ou customize as linhas e os marcadores, s√£o muitos tra√ßos para encontrar facilmente uma inst√¢ncia desejada. Assumindo que o objetivo da visualiza√ß√£o √© destacar os dados Brasileiros dentro da Am√©rica do Sul, ter controle sobre cada tra√ßo individualmente facilita a customiza√ß√£o de um elemento espec√≠fico.
> N√£o ficou t√£o bonito, mas √© s√≥ um exemplo n√©? ü§ó

Usando Graph Objects voc√™ tamb√©m se habitua a usar m√©todos essenciais como o `update_layout`, nosso pr√≥ximo assunto :)

### Personaliza√ß√£o B√°sica
Usando plotly express, muitos recursos de personaliza√ß√£o est√£o dispon√≠veis como par√¢metros da fun√ß√£o de plot. Vale a pena fu√ßar a documenta√ß√£o do plot que voc√™ deseja para explorar as customiza√ß√µes dispon√≠veis (por exemplo: [documenta√ß√£o do plot de barras](https://plotly.com/python-api-reference/generated/plotly.express.bar.html)).

````python
fig = px.bar(df[df.country=='Brazil'], x='year', y='pop', color='lifeExp',
             title ="Popula√ß√£o do Brasil ao longo dos anos", width=800,
             ## quais dados estar√£o no mouse hover?
             hover_data=['lifeExp', 'gdpPercap'],
             ## qual o apelido das dimens√µes?
             labels={'pop':'Popula√ß√£o',
                     'lifeExp': 'Exp. de vida',
                     'gdpPercap': 'PIB per capita',
                     'year': 'Ano',},

)

fig.show()
````

Por√©m, o m√©todo `update_layout` de objetos `Figure` (veja a [documenta√ß√£o](https://plotly.com/python/reference/layout/)) oferece possibilidades muito mais completas para personaliza√ß√£o da nossa figura, incluindo fonte, espa√ßamento, configura√ß√µes de legenda, etc. Em casos em que trabalhamos diretamente com Graph Objects (sem usar o Plotly Express) √© atrav√©s desse m√©todo que definimos at√© mesmo as coisas mais simples.

Em ambos os casos (fun√ß√µes do `px` ou `update_layout`), o Plotly oferece um facilitador de sintaxe interessante. Os atributos podem ser configurados de duas maneiras:
* sintaxe de dicion√°rio: os atributos s√£o internamente organizados em dicion√°rios que podemos acessar diretamente.
* `snake_case`: O uso do caracter `_` (underline) para agregar chaves dos dicion√°rios.

````python
fig = px.bar(df[df.country=='Brazil'], x='year', y='pop', color='lifeExp', width=800)

# sintaxe de dicion√°rio
fig.update_layout(
    title="Popula√ß√£o do Brasil ao longo dos anos",
    xaxis={'title': "Ano"},
    yaxis={'title': "Popula√ß√£o"},
    font=dict(
        family="Courier New, monospace",
        size=16,
        color="RebeccaPurple"
    )
)

# sintaxe snake_case
fig.update_layout(
    title="Popula√ß√£o do Brasil ao longo dos anos",
    xaxis_title = "Ano",
    yaxis_title = "Popula√ß√£o",
    font_family="Courier New, monospace",
    font_size=16,
    font_color="RebeccaPurple"
)

fig.show()
````

### Subplots

O Plotly Express oferece recursos que facilitam a cria√ß√£o de m√∫ltiplos plots em uma mesma figura, incluindo a adi√ß√£o de gr√°ficos marginais utilizando atributos como marginal_x. Esses recursos s√£o particularmente √∫teis em gr√°ficos de distribui√ß√£o e dispers√£o (scatter), permitindo visualizar dados adicionais.

````python
df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
           marginal_y="violin", marginal_x="box", template="ggplot2", width=800)
fig.show()
````

Tamb√©m temos o j√° conhecido `facet`, nossos **pequenos m√∫ltiplots**, tamb√©m dispon√≠veis no Plotly Express. Nessa visualiza√ß√£o, plotamos o mesmo dado dividido em subgrupos. Os atributos s√£o:
* `facet_row`: Recebe uma coluna do DataFrame para organizar os gr√°ficos em m√∫ltiplas linhas, cada uma correspondendo a uma categoria ou valor √∫nico da vari√°vel.
* `facet_col`: Recebe uma coluna do DataFrame para organizar os gr√°ficos em m√∫ltiplas colunas, cada uma correspondendo a uma categoria ou valor √∫nico da vari√°vel.
* `facet_col_wrap`: Define o n√∫mero m√°ximo de colunas, quebrando em novas linhas e criando um grid de visualiza√ß√µes.

````python
df = px.data.gapminder().query("year==2007")
fig = px.scatter(df, x="gdpPercap", y="lifeExp", size="pop", hover_name="country",
                facet_col="continent", color="continent",
                log_x=True, size_max=45, range_x=[100,100000], range_y=[25,90])
fig.show()
````

Mas a forma gen√©rica de criar subplots √© atrav√©s do m√≥dulo `subplots` do Plotly, mais especificamente sua fun√ß√£o `make_subplots`. Ela funciona de maneira similar ao que aprendemos com Graph Objects, ou seja, adicionando novos tra√ßos, por√©m cada tra√ßo possui um indicador da linha e da coluna onde deve ser visualizado.

````python
from plotly.subplots import make_subplots

df = px.data.iris()

# Criando subplots
fig = make_subplots(rows=1, cols=2, subplot_titles=("Gr√°fico 1", "Gr√°fico 2"))

# Adicionando tra√ßos aos subplots
fig.add_trace(go.Scatter(x=df['sepal_width'], y=df['sepal_length'], mode='markers'), row=1, col=1)
fig.add_trace(go.Scatter(x=df['petal_width'], y=df['petal_length'], mode='markers'), row=1, col=2)

fig.update_layout(title="Subplots Exemplo", width=900, showlegend=False)
fig.show()
````

````python
import pandas as pd

df = pd.read_csv(
    "https://raw.githubusercontent.com/plotly/datasets/master/volcano_db.csv",
    encoding="iso-8859-1",
)
freq = df.Country.value_counts().reset_index()
df_v = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/volcano.csv")

## Criando a figura com subplots
fig = make_subplots(
    rows=1, cols=2,
    column_widths=[0.5, 0.5],

    ## por padr√£o considera-se um tipo xy de gr√°fico
    specs=[[{"type": "scattergeo"}, {"type": "bar"}]]
)

# Add scattergeo globe map of volcano locations
fig.add_trace(
    go.Scattergeo(lat=df["Latitude"],
                  lon=df["Longitude"],
                  mode="markers",
                  hoverinfo="text",
                  showlegend=False,
                  marker=dict(color="crimson", size=4, opacity=0.8)),
    row=1, col=1
)

# Update geo subplot properties
fig.update_geos(
    projection_type="orthographic",
    landcolor="white",
    oceancolor="MidnightBlue",
    showocean=True,
    lakecolor="LightBlue"
)

# Add locations bar chart
fig.add_trace(
    go.Bar(y=freq["count"][0:10],x=freq["Country"][0:10], marker=dict(color="crimson"), showlegend=False),
    row=1, col=2
)
fig.update_xaxes(tickangle=45)

fig.update_layout(height=400, width=800)
fig.show()
````

#### Bot√µes e Menu Dropdown

Ambos os recursos de bot√£o e menu dropdown (ou drop em qualquer dire√ß√£o) usam do par√¢metro `updatemenus` do m√©todo `update_layout`. Seu par√¢metro √© uma lista de dicion√°rios, onde cada dicion√°rio √© um menu que se deseja adicionar. O que distingue um menu dropdown de um bot√£o √© o atributo `type` desse dicion√°rio.

Cada bot√£o tem um par√¢metro chamado `method` que define o tipo de altera√ß√£o que ser√° feita no gr√°fico. Suas op√ß√µes s√£o:

* "**restyle**": modificar dados ou atributos de dados
* "**relayout**": modificar atributos de layout
* "**update**": modificar dados e atributos de layout; combina√ß√£o de "restyle" e "relayout"


> Tutorial bot√µes: https://plotly.com/python/custom-buttons/ <br>
Tutorial dropdown menu: https://plotly.com/python/dropdowns/

````python
df = px.data.gapminder()
fig = px.bar(df[df.country=='Argentina'], x='year', y='pop', color='lifeExp',
             title="Popula√ß√£o ao longo dos anos: Argentina",)

# sintaxe de dicion√°rio
fig.update_layout(
    xaxis={'title': "Ano"},
    yaxis={'title': "Popula√ß√£o"},
    font=dict(
        family="Courier New, monospace",
        size=16,
        color="RebeccaPurple"
    )
)

# Adicionar bot√£o de interatividade para alterar o pa√≠s
fig.update_layout(
    updatemenus=[
        ### Menu dropdown de pa√≠ses
        dict(
            direction="down",
            buttons=list([
                dict(
                    args=[
                    # dicion√°rio de altera√ß√µes de dados
                    {"x": [df[df['country']==country]['year']],
                    "y": [df[df['country'] == country]['pop']],
                    },
                    # dicion√°rio de altera√ß√µes de layout
                    {"title": f"Popula√ß√£o ao longo dos anos: {country}"
                    }],
                    label=str(country),
                    method="update"
                ) for country in sa_countries
            ]),
            showactive=True,
        ),

        ### Bot√µes de tipo de gr√°fico
        dict(
            type="buttons",
            direction='down',
            buttons=list([
                dict(
                    args=[{"type": grafico}],
                    label=grafico,
                    method='restyle',
                ) for grafico in ['bar', 'scatter']
            ]),
            showactive=True,
            x=-0.07,
            y=0.85

        )
    ],
    # Rangeslider
    xaxis=dict(
        rangeslider=dict(
            visible=True
        ),
    ),
    height=600, width=900
)
````
